```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = tf.range(-100, 100, 4)
X
```

```python
y = X + 10
y
```

```python
plt.figure(figsize=(5, 3))
plt.scatter(X, y)
plt.show()
```

```python
# split to training data & test data
X_train = X[:40]
y_train = y[:40]

X_test = X[40:]
y_test = y[40:]
```

```python
# plot training data & test data
plt.figure(figsize=(5, 3))
plt.scatter(X_train, y_train, c='b', label='training data')
plt.scatter(X_test, y_test, c='g', label='test data')
plt.legend()
plt.show()
```

```python
# create a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation='relu', input_shape=[1], name="hidden_1"),
    tf.keras.layers.Dense(100, activation='relu', name="hidden_2"),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(1, name="output")
])

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(lr=0.0001),
              metrics=["mae"]
             )

model.fit(X_train, y_train, epochs=1000, verbose=1)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model, show_shapes=True)
```

```python
# create a plotting function
def plot_predictions(training_data=X_train,
                    training_label=y_train,
                    test_data=X_test,
                    test_label=y_test,
                    prediction=y_pred):
    '''
    Plot training data, test data, and compare predictions to ground truth.
    '''
    plt.figure(figsize=(5, 3))
    plt.scatter(training_data, training_label, c='b', label='training data')
    plt.scatter(test_data, test_label, c='g', label='test ground truth')
    plt.scatter(test_data, prediction, c='r', label='test prediction')
    plt.legend()
    plt.show()
```

```python
plot_predictions()
```

```python
model.evaluate(X_test, y_test)
```

```python
mae = tf.metrics.mean_absolute_error(y_true=y_test,
                                    y_pred=tf.squeeze(y_pred))
mae.numpy()
```

```python
mse = tf.metrics.mean_squared_error(y_true=y_test,
                                   y_pred=tf.squeeze(y_pred))
mse.numpy()
```

## Comparision of the results (of 3 models)

```python
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model_1.compile(loss=tf.keras.losses.mae,
               optimizer=tf.keras.optimizers.SGD(),
               metrics=["mae"])

model_1.fit(X_train, y_train, epochs=10, verbose=0)
```

```python
y_pred_1 = model_1.predict(X_test)
```

```python
mae_1 = tf.metrics.mean_absolute_error(y_true=y_test,
                                      y_pred=tf.squeeze(y_pred_1))
mse_1 = tf.metrics.mean_squared_error(y_true=y_test,
                                     y_pred=tf.squeeze(y_pred_1))
```

```python
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(1)
])

model_2.compile(loss=tf.keras.losses.mae,
               optimizer=tf.keras.optimizers.SGD(),
               metrics=["mae"])

model_2.fit(X_train, y_train, epochs=10, verbose=0)
```

```python
y_pred_2 = model_2.predict(X_test)
```

```python
mae_2 = tf.metrics.mean_absolute_error(y_true=y_test,
                                      y_pred=tf.squeeze(y_pred_2))
mse_2 = tf.metrics.mean_squared_error(y_true=y_test,
                                     y_pred=tf.squeeze(y_pred_2))
```

```python
model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model_3.compile(loss=tf.keras.losses.mae,
               optimizer=tf.keras.optimizers.Adam(lr=0.001),
               metrics=["mae"])

model_3.fit(X_train, y_train, epochs=100, verbose=0)
```

```python
y_pred_3 = model_3.predict(X_test)
```

```python
mae_3 = tf.metrics.mean_absolute_error(y_true=y_test,
                                      y_pred=tf.squeeze(y_pred_3))
mse_3 = tf.metrics.mean_squared_error(y_true=y_test,
                                     y_pred=tf.squeeze(y_pred_3))
```

```python
results = [['model_1', mae_1.numpy(), mse_1.numpy()],
          ['model_2', mae_2.numpy(), mse_2.numpy()],
          ['model_3', mae_3.numpy(), mse_3.numpy()]]
```
