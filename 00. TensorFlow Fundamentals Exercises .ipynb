1. Create a vector, scalar, matrix and tensor with values of your choosing using tf.constant().

```python
import tensorflow as tf
```

```python
scalar = tf.constant(7)
print(scalar)
```

```python
vector = tf.constant([8, 3])
print(vector)
```

```python
matrix = tf.constant([[3, 4],
                     [5, 3]])
print(matrix)
```

```python
tensor = tf.constant([[[3, 4, 5],
                      [3, 2, 1]],
                      
                     [[7, 4, 2],
                     [2, 2, 3]]])
print(tensor)
```

2. Find the shape, rank and size of the tensors you created in 1.

```python
tensor.shape
```

```python
tensor.ndim
```

```python
tf.size(tensor)
```

3. Create two tensors containing random values between 0 and 1 with shape [5, 300].

```python
tensor_1 = tf.random.uniform((5, 300))
print(tensor_1)
```

```python
tf.random.set_seed(1234)
tensor_2 = tf.random.uniform((5, 300), seed=3)
print(tensor_2)
```

4. Multiply the two tensors you created in 3 using matrix multiplication.

```python
tf.tensordot(tensor_1, tensor_2, axes=[1, 1])
```

```python
tf.matmul(tensor_1, tf.transpose(tensor_2))
```

6. Find the min and max values of the tensor you created in 5 along the first axis.
